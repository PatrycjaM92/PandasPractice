```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2,2,3000)

fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.axhline(y=0,color='black', linewidth=1)
ax.axvline(x=0,color='black', linewidth=1)
ax.plot(x,3*x+1,color="r")  
ax.plot(x,x**2+2,color="green")
ax.plot(x,x**3+3,color="purple")
ax.grid(True)
plt.xticks(np.arange(-3, 3.5, step=0.5),rotation=45)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Linear and quadratic",fontdict={"family":"Comic Sans MS","size":"12","color":"blue"})
```

### BEP - Break-Event-Point <br> total Cost = total Income
#####  Próg rentowności ilościowy = koszty stałe  /  (Cena - Koszty zmienne jdn)

```python
# Koszty stałe w zł
kS = 3000 
# Cena za jednostkę produktu zł
c = 100
# Koszt zmienny na jednostkę produktu w zł
kZ= 50

BEP_ilosciowy = kS/(c-kZ)
BEP_ilosciowy
```

### BEP wartościowy
#### Próg rentowności warościowy = Cena_jnd * Koszty_stale / (Cena - Koszt zmienny jedn)

```python
BEP_wartosciowy = c*kS/(c-kZ)
BEP_wartosciowy
```

```python
ks=3000
kzj=20
ilosc=np.arange(0,200)
kc=ilosc*kzj+ks
c=60
ks_values=[ks]*len(ilosc)
fig, ax = plt.subplots()  # Create a figure containing a single axes.
ax.plot(ilosc,kc,color="r",label="Koszt całkowity")  
ax.plot(ilosc,c*ilosc,color="green",label="Sprzedaż w zł")
ax.plot(ilosc,ks_values,color='blue',label="Koszty stałe")
ax.grid(True)
#plt.xticks(np.arange(0, 200, step=25),rotation=45)
plt.xlabel("Spzedaż (szt)")
plt.ylabel("Przychody,Koszty")
plt.legend()
plt.title("BEP",fontdict={"family":"Comic Sans MS","size":"12","color":"blue"})

ax.annotate('Koszt Stały', xy=(100, 0), xytext=(150, 3050))
print(ks/(c-kzj))
```
